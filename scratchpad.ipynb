## Wonder if we can graph the data easily

```python
import pandas as pd
from matplotlib import pyplot as plt
import src.acquire
```

```python
request = 'https://api.coindesk.com/v1/bpi/historical/close.json'
```

```python
data = src.acquire.get_bit_data(request)
```

```python
df = pd.DataFrame.from_dict(data, orient='index')
```

```python
df.head()
```

```python
df.plot(figsize=[16, 9],
        title="Bitcoin Prices, Last 31 Days",
        grid=True)
```
