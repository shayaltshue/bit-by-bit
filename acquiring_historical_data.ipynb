```python
import requests
```

```python
response = requests.get("https://api.coindesk.com/v1/bpi/historical/close.json")
```

```python
response.text
```

The API returns a string containing JSON data. We'll need the built-in json library to help us handle all this data.

```python
import json
```

```python
data = json.loads(response.text)
```

Our data is stored under BPI. We can use a key to access it. We'll save that data into a new variable called 'bpi_data'

```python
bpi_data = data['bpi']
bpi_data
```
